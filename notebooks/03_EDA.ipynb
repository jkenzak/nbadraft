# 03 Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed/merged_data.csv')
quant_vars = df.select_dtypes(include=['number']).columns[1:] # Exclude Draft Yr
for quant_var in quant_vars:
    plt.figure(figsize=(8, 6))
    grouped_data = df.groupby('All-Star')[quant_var].mean()
    grouped_data.plot(kind='bar')
    plt.title(f'{quant_var} by All-Star')
    plt.xlabel('All-Star')
    plt.ylabel(f'Mean {quant_var}')
    plt.show()
```
