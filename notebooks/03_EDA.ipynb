# Exploratory Data Analysis (EDA) - NBA Draft Prediction Project
This notebook explores trends in NCAA player stats and NBA Combine data. We also compare future All-Stars with non-All-Stars to identify predictive signals.

# 03 Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset Overview
We begin by loading the cleaned dataset and viewing basic statistics and structure.

```python
df = pd.read_csv('../data/processed/merged_data.csv')
quant_vars = df.select_dtypes(include=['number']).columns[1:] # Exclude Draft Yr
for quant_var in quant_vars:
    df.boxplot(column=quant_var, by='All-Star', grid=False)
    plt.title(f'{quant_var} Distribution by All-Star')
    plt.suptitle('')
    plt.xlabel('All-Star')
    plt.ylabel(quant_var)
    plt.show()
```

## EDA Complete
Weve gained insights into the key differences between future All-Stars and other players. The next step is to train a predictive model based on these patterns in `04_modeling.ipynb`.
